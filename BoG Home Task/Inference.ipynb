```python
def model_inference(data, model_path = 'best_xgb_model.pkl'):
    
    with open('best_xgb_model.pkl', 'rb') as f:
        xgb = pickle.load(f)
    
    if isinstance(df, pd.Series):
        data = df.to_frame().T
    elif isinstance(data, dict):
        data = pd.DataFrame([data])
    
    clean_df = clean_data(data)
    
    try:
        clean_df = clean_df.drop(['is_canceled'], axis = 1)
    except:
        pass
    
    pred = xgb.predict(clean_df)
    
    return pred
```

```python
for i in range(1,20):
    
    bookings = pd.read_csv('hotel_bookings.csv')
    sample_bookings = bookings.sample(n=1000)

    predictions = model_inference(sample_bookings.copy())
    actual = clean_data(sample_bookings)['is_canceled']
    print(f'accuracy of run {i}:',(predictions == actual).sum()/len(predictions))
```
